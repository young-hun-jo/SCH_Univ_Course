```python
import pandas as pd
import numpy as np

# 다른 디릭토리에 있는 2개 파일 디렉토리 각각 list에 담기
data_dirs = ['../data/murphy/learn', '../data/murphy/test']

# 합칠 빈 데이터프레임 생성
original_data = pd.DataFrame()

original_data.shape
```

```python
import os

# 위에서 정의해준 여러 디렉토리를 반복문을 통해서 각 디렉토리에 있는 파일들 읽어오기
for dd in data_dirs:
    filenames = os.listdir(dd)
    # sort해야 나중에 디버깅하거나 케이스분석하기 편함!
    filenames.sort()
    print(f"{dd} 안에는 {len(filenames)}개의 파일이 존재")
    print(filenames[:10])
    
# 리스트[] 안에 데이터프레임도 들어갈 수 있음!! 리스트는 메우 flexible
```

```python
# 파일 6000개의 구조가 똑같이 생겼으니 1개의 파일로 만드는 작업을 하자!
# 데이터 번호(파일 번호)도 추가로 넣어주기

targetfilepath = '../data/prac002.data'
for dd in data_dirs:
    filenames = os.listdir(dd)
    filenames.sort()
    print(f"{dd} 안에는 {len(filenames)}개의 파일이 존재")
    for fn in filenames:
        # 각 디렉토리안의 여러가지 파일들 하나씩 읽어나가기
        full_path = dd + '/' + fn
        with open(full_path, 'r') as rf:
            #모든 줄을 읽어서 출력하자.
            while True:
                content = rf.readline()
                # content가 없다면 = 더이상 읽을 줄이 없다면 멈추기
                if not content : break
                while '  ' in content:
                    # 두 개의 공백을 하나의 공백으로 모두 줄여주자
                    content = content.replace('  ', ' ')
                
                # 'a' 모드 : 파일의 마지막 내용에 추가작성하는 모드(1번만 실행!)
                with open(targetfilepath, 'a') as wf:
                    wf.write(content)
        
```

```python
original_data = pd.read_csv(targetfilepath, sep=' ', dtype='unicode',
                           names=['class','objnum','type','xx1',
                                 'yy1','xx2','yy2','size','diag'])
print(original_data.shape)
original_data.head()
```

```python
# class를 맞춰주기
# class별로 objnum이 몇개 있는지 봐보자
pd.crosstab(original_data['class'], original_data.objnum, margins=True)
```
