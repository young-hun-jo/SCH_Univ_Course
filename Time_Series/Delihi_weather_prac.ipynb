```r
train$year = lubridate::year(train$date)
train$month = lubridate::month(train$date)
train
```

```r
library(dplyr)
```

```r
train_df<-train %>% group_by(year, month) %>% summarise(avg=mean(meantemp)) %>% as.data.frame
head(train_df, n=3)
```

```r
library(tidyverse)
```

```r
train_month<-train_df %>% spread(key='month', value='avg')
train_month<-train_month[-nrow(train_month),]
```
