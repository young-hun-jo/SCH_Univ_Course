```r
g<-read.csv('군집분석.csv', head=T)
head(g)
```

```r
g<-graph.data.frame(g, directed = T)
plot(g)
```

```r
plot(g, layout=layout.kamada.kawai, vertex.size=0.5, edge.color='red',
    edge.arrow.size=0.3, edge.arrow.width=0.1)
```

```r
library(stringr)
gubun<-V(g)$name
gubun<-str_sub(gubun, start=1, end=1)
color<-ifelse(gubun=='S', 'red','blue')
size<-ifelse(gubun=='S', 1, 6)
plot(g, layout=layout.fruchterman.reingold, vertex.size=size, vertex.color=color,
    edge.arrow.width=0.2, edge.arrow.size=0.5)
```

```r
shapes<-ifelse(gubun=='S','square','circle')
plot(g, layout=layout.fruchterman.reingold,
    vertex.shape=shapes, vertex.color=color)
```
