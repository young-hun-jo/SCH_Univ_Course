- tibble 데이터타입 = tidyverse에서 제공하는 데이터 타입

```r
setwd("C:/Users/joyh1/Desktop/빅데이터_20-1/R_data")
flights<-read.csv("flights.csv")
```

```r
# read_csv인자에 col_names=벡터 로 칼럼이름 설정 가능
# read_csv인자에 na='.' 이라고하면 . 을 NA로 출력시켜줌
head(flights, n=5)
```

```r
# . => special placeholder라는 의미로 '앞의 데이터'를 의미!!
flights %>% .$AIRLINE

flights %>% .[['DAY']]

# flights[1] => dataframe으로출력
# lfights[[1]] => vector로 출력
```

```r
# 그룹핑하기 => group_by 사용
by_dest<- flights %>% group_by(., DESTINATION_AIRPORT)
head(by_dest, n=4)
```

```r
# summarise 사용해 집계하기
# summarise(df, new_column= ~, new_column=~, )
delay<-summarise(by_dest, count=n(), dist=mean(DISTANCE, na.rm=TRUE), delay=mean(ARRIVAL_DELAY, na.rm=TRUE))
head(delay, n=5)
```

```r
# 변수추출1 -> select함수
head(select(flights, contains("TIME")), n=5)
# 해당 df에서 'TIME'을 포함하는 칼럼들만 추출해라
```

```r
# 변수추출2
head(select(flights, YEAR:DAY, ends_with('DELAY')),n=5)
# YEAmR~DAY칼럼과 DELAY로 끝나는 칼럼명들 추출해라
```

```r
# 변수추출3 -> filter
head(flights %>% filter(between(ARRIVAL_DELAY,0,60)), n=5)

# ARRIVAL_DELAY값이 0~60사이의 값을 출력해라
```

# 기술통계량

## 통계적특질 

- 중심화 경향(대표값) 
    * 산술평균, 기하평균, 조화평균, 가중평균, 중앙값, 최빈값
- 퍼짐 정도
    * 분산, 표준편차, 변이계수, 범위, IQR(3분위수-1분위수), 백분위수
- 분포형태와 대칭정도
    * 왜도(왼쪽,오른쪽 비대칭), 첨도(봉우리가 첨예한지)
 
- 3가지 종류 종합적으로 보아야 함
- 특질 추출 전 summary와 그래프 간단 시각화해 살펴보는것이 바람직

- 사용할 패키지
    * seewave
    * signal
    * pracma
    * fbasics

```r
library(fBasics)
```

```r
x<-sample(1000)

# 산술평균
mean(x)

# 중앙값
median(x)

# 최빈값 -> table화시킬수있음
which.max(table(x))

# 기하평균
prod(x)^(1/1000)

# 조화평균
1/mean(1/x)

# 분산
var(x)

#표준편차
sd(x)

#변이계수
100*sd(x)/mean(x)

#범위(range)
diff(range())

#IQR
IQR(x)

#최소값
min(x)

#최대값
max(x)

#백분위수
quantile(x)

#왜도 => fBasics 패키지에 있음
skewness(x) 

#첨도 => fBasics 패키지에 있음
kurtosis(x)
```

```r
# 분위수 - 분위수
qqnorm(x)
```
