```r
# 가속도 크기(magnitude) 구하기

# 가속도 구해서 새로운 칼럼에 추가하는 함수 만들기
mag<-function(df,column){
    # with( )는 데이터 프레임 또는 리스트 내 필드를 필드 이름만으로 접근할 수 있게 해주는 함수다.
    # get() 은 get인자안의 변수에 대한 값들을 출력한다.
    df[,str_c("mag",column)]<-with(df,sqrt(get(str_c(column,'.x'))^2+
                                          get(str_c(column, '.y'))^2+
                                          get(str_c(column, '.z'))^2))
    return(df)
}
```

```r
library(dplyr)
```

```r
# user1의 walking 데이터 넣기위해 새로운 df생성
user1_walking_total<-data.frame()

# user1의 데이터 반복문돌려서 입력
for (f in user1_walking){
    #get함수를 이용해서 해당 변수에 할당되어있는 데이터프레임값 할당
    temp<-get(f)
    # 만들었던 새로운 df에 새로운 행(r)붙이기
    # mutate(df명, new변수명=칼럼연산(기타연산))
    user1_walking_total<-rbind(user1_walking_total, temp %>%
                              mutate(exp_no=unlist(regmatches(f,gregexpr("[[:digit:]]+",f)[1]))[1],
                                                  id=unlist(regmatches(f,gregexpr("[[:digit:]]+",f)[1]))[2]))
}


# [[:digit:]] => 숫자인것만을 나타냄
# gregexpr("졍규표현식=패턴", 벡터or문자) => 해당 위치를 반환
# regmatches(벡터or문자, 패턴이있는 위치) => 패턴을 반환
# exp_no, id 라는 새로운 칼럼을 mutate함수를 통해서 추가해줌
```

```r
# mag함수를 통해서 magnitude(가속도)칼럼 추가해주기
# 가속도칼럼 이름은 "maguserAcceleration"
user1_walking_total<-mag(user1_walking_total, "userAcceleration")
head(user1_walking_total, n=3)
```

```r
# 새로만들었던 exp_no 별로 시간값을 생성. 시간값은 데이터 순번으로 하기
# exp_no으로 그룹핑했다가 다시 언그룹으로 풀기
user1_walking_total<-user1_walking_total %>% group_by(exp_no) %>% mutate(time=row_number()) %>% ungroup()
head(user1_walking_total, n=4)
```

```r
# user1의 세가지 경우(7,8,15)일때 각각 '시간에 따른 가속도' 그래프 비교 시각화
# facet_wrap(.~exp_no) => exp_no별로 subplot그리겠다!
# facet_wrap(nrow=3) => 그래프를 3개의 행으로 그리겠다!
ggplot(user1_walking_total, aes(x=time, y=maguserAcceleration))+geom_line()+facet_wrap(.~exp_no,nrow=3)
```

```r
# user1만이 아닌 전체데이터로 만들기
HAR_total<-data.frame()

for (f in fls){
    temp<-get(f)
    # 전체데이터를 담을 df에 행(r)추가
    HAR_total<-rbind(HAR_total, temp %>%
                    mutate(exp_no=unlist(regmatches(f,gregexpr("[[:digit::]]+",f)[1]))[1],
                          id=unlist(regmatches(f,gregexpr("[[:digit:]]+",f)[1]))[2],
                          acitivity=unlist(str_split(f,"\\_"))[1]))
}

# 기존에 하던 거랑 똑같이하고 어떤 '동작'을 나타내는지 표시하기 위해 새로운 'activity'칼럼 추가
```

```r
# 가속도(magnitude), 회전(rotationRate) 새로운 칼럼 만들기
HAR_total<-mag(HAR_total,"userAcceleration")
HAR_total<-mag(HAR_total,"rotationRate")
```

```r
head(HAR_total, n=4)
```

```r
# 기술통계랑 특질 추출하기

# 왜도(skewness) 함수만들기
skewness<-function(x){
    (sum((x-mean(x))^3)/length(x))/
    ((sum((x-mean(x))^2)/length(x)))^(3/2)
}
```

```r
# 가속도와 회전 칼럼에 대한 특질 생성
# summarise_at : 선택한 변수에 대해서 여러 함수를 적용할 때 사용!

HAR_summary<-HAR_total %>% group_by(id,exp_no,acitivity) %>%
summarize_at(.vars=c("maguserAcceleration","magrotationRate"),.funs=c(mean,min,max,sd,skewness))
# .vars => 앞의 데이터의 변수라는 뜻, .funs => 앞의 데이터에 적용할 함수들, 함수적용할때는 따옴표 안붙임!
```
