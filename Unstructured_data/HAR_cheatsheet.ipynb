```r
# 가속도 크기(magnitude) 구하기

# 가속도 구해서 새로운 칼럼에 추가하는 함수 만들기
mag<-function(df,column){
    # with( )는 데이터 프레임 또는 리스트 내 필드를 필드 이름만으로 접근할 수 있게 해주는 함수다.
    # get() 은 get인자안의 변수에 대한 값들을 출력한다.
    df[,str_c("mag",column)]<-with(df,sqrt(get(str_c(column,'.x'))^2+
                                          get(str_c(column, '.y'))^2+
                                          get(str_c(column, '.z'))^2))
    return(df)
}
```

```r
# user1의 walking 데이터 넣기위해 새로운 df생성
user1_walking_total<-data.frame()

# user1의 데이터 반복문돌려서 입력
for (f in user1_walking){
    #get함수를 이용해서 해당 변수에 할당되어있는 데이터프레임값 할당
    temp<-get(f)
    # 만들었던 새로운 df에 새로운 행(r)붙이기
    # mutate(df명, new변수명=칼럼연산(기타연산))
    user1_walking_total<-rbind(user1_walking_total, temp %>%
                              mutate(exp_no=unlist(regmatches(f,gregexpr("[[:digit:]]+",f)[1]))[1],
                                                  id=unlist(regmatches(f,gregexpr("[[:digit:]]+",f)[1]))[2]))
}


# [[:digit:]] => 숫자인것만을 나타냄
# gregexpr("졍규표현식=패턴", 벡터or문자) => 해당 위치를 반환
# regmatches(벡터or문자, 패턴이있는 위치) => 패턴을 반환
# exp_no, id 라는 새로운 칼럼을 mutate함수를 통해서 추가해줌
```

```r
# 기술통계랑 특질 추출하기

# 왜도(skewness) 함수만들기
skewness<-function(x){
    (sum((x-mean(x))^3)/length(x))/
    ((sum((x-mean(x))^2)/length(x)))^(3/2)
}
```

```r
# 가속도와 회전 칼럼에 대한 특질 생성
# summarise_at : 선택한 변수에 대해서 여러 함수를 적용할 때 사용!

HAR_summary<-HAR_total %>% group_by(id,exp_no,acitivity) %>%
summarize_at(.vars=c("maguserAcceleration","magrotationRate"),.funs=c(mean,min,max,sd,skewness))
# .vars => 앞의 데이터의 변수라는 뜻, .funs => 앞의 데이터에 적용할 함수들, 함수적용할때는 따옴표 안붙임!
```

# 변화분석하기
## 변화분석 특징 추출 ->changepoint 라이브러리
### library(changepoint) 로드
    
    - cpt.mean : 변화들만을 평균내는것 = 간격의 평균값
    - cpt.var : 변화들만의 분산을 내는것 = 간격의 분산값
    - cpt.meanvar : 변화들만의 평균과분산을 내는 것 = 평균,분산변화가 얼마나 일어나는지

```r
fls<-ls()[str_detect(ls(),'.csv')]
total_peak<-data.frame()
```

### 변화 횟수 변수화

#### cpts : 몇번째에서 변화량이 발생했는지 알려주는 함수!

```r
for (d in fls){
    f<-get(d)
    f<-mag(f,"rotationRate")
    f<-mag(f,'userAcceleration')
    # 가속도와 회전율 칼럼을 뽑아서 거기다가 cpt.mean함수 적용한 값들을 rslt에 담기
    rslt<-sapply(f%>%select(magrotationRate,maguserAcceleration), cpt.mean)
    # 위에서만든 rslt에 가속도랑회전 cpt.mean값있는데 거기서 몇번째에 변화량이 발생했는지 담기
    rslt_cpts1<-cpts(rslt$magrotationRate)
    rslt_cpts2<-cpts(rslt$maguserAcceleration)
    
    rslt2<-sapply(f%>%select(magrotationRate,maguserAcceleration), cpt.var)
    rslt2_cpts1<-cpts(rslt$magrotationRate)
    rslt2_cpts2<-cpts(rslt$maguserAcceleration)
    
    rslt3<-sapply(f%>%select(magrotationRate,maguserAcceleration), cpt.meanvar)
    rslt3_cpts1<-cpts(rslt$magrotationRate)
    rslt3_cpts2<-cpts(rslt$maguserAcceleration)
    
    # total_peak 데이터프레임에 rbind로 위에서 만든 
    total_peak<-rbind(total_peak, data.frame(d, cp1=length(rslt_cpts1), cp2=length(rslt_cpts2),
                                            cp3=length(rslt2_cpts1), cp4=length(rslt2_cpts2), cp5=length(rslt3_cpts1),
                                            cp6=length(rslt3_cpts2)))
}
```
